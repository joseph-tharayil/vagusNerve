```python
import matplotlib
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
```

```python
def getTime():

    nx=500000

    tmin=-3 # In s
    tmax=3 # In s
    time=np.arange(tmin,tmax,(tmax-tmin)/(nx-1))

    return time
```

```python
time = np.linspace(0,2,10000)
```

```python
a = np.load(r'D:\vagusNerve\SFAPValidation\analytic.npy')
s = np.load(r'D:\vagusNerve\SFAPValidation\sensing.npy')
```

```python
tidx = np.where(time>0.3)[0]
```

```python
plt.figure(figsize=(6,3))
plt.subplot(121)
plt.plot(time,a,'-')
plt.plot(time,s,'-')
plt.xlabel('Time (ms)')
plt.ylabel("SFAP (V)")
plt.legend(['Analytic Signal','S4L Sensing'])

plt.subplot(122)
plt.scatter(a[tidx],s[tidx])
plt.plot([-6e-7,5e-7],[-6e-7,5e-7],'k')
plt.xlabel('Analytic\nSignal (V)')
plt.ylabel('S4L Sensing (V)')

plt.tight_layout()
plt.savefig('s4l_validation.png',dpi=600)
```
