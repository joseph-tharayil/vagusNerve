```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
a = np.load(r'D:\vagusNerve\SFAPValidation\analytic.npy')
p = np.load(r'D:\vagusNerve\SFAPValidation\pointSource.npy')
```

```python
time = np.linspace(0,30,1000)
tidx = np.intersect1d(np.where(time>4),np.where(time<12))
```

```python
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.plot(time[tidx],a[tidx])
plt.plot(time[tidx],p[tidx])
plt.xlabel('Time (ms)')
plt.ylabel("SFAP (V)")
plt.legend(['Analytic\nSignal','Point-source\napproximation'],loc='lower left')

plt.subplot(1,2,2)
plt.scatter(a[tidx],p[tidx])
plt.plot([-3.5e-10,.5e-10],[-3.5e-10,.5e-10],'k')
plt.xlabel('Analytic Signal (V)')
plt.ylabel('Point-source\nApproximation (V)')

plt.tight_layout()
plt.show()
plt.savefig('pointsource_validation.png',dpi=600)
```

```python
from scipy.stats import linregress
reg = linregress(a[tidx],p[tidx])
print(reg)
```
