```python
import numpy as np
```

```python
frequencies = np.array([100,200,300,400,500])
```

```python
times = np.array([0.0,0.09,0.1,0.29,0.3,0.49,0.5])
currents = np.array([0,0,-1,-1,1,1,0])*0.7#*1.45312#
```

```python
intervals = 1/frequencies*1e3 # Inter-stimulus intervals, in ms
```

```python
numRepetitions = 10
```

```python
for j, interval in enumerate(intervals):
    newtimes = times
    newcurrents = currents
    for i in range(numRepetitions):
        newtimes = np.hstack((newtimes,times+interval*(i+1)))
        newcurrents = np.hstack((newcurrents,currents))

    output = np.vstack((newtimes,newcurrents))

    np.savetxt('stimulus_'+str(frequencies[j])+'_M.txt',output)
```

```python
import matplotlib.pyplot as plt
%matplotlib widget

numAPs_10um = np.array([12,12,6,6,6])/12
numAPs_4um = np.array([11,11,6,6,6])/11
plt.figure(figsize=(4,4))
plt.plot(frequencies,numAPs_10um,'o',alpha=0.25)
plt.plot(frequencies,numAPs_4um,'o',alpha=0.25)
plt.xlabel('Pulse repetition frequency (Hz)')
plt.ylabel('Fraction of pulses triggering APs')
plt.legend(['10 $\mu$m fiber','4 $\mu$m fiber'])
plt.tight_layout()
plt.savefig('pulseRepetition.png',dpi=600)
```
